```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0
b = 2
N = 100
```

```python
dx = (b - a) / (N-1)
```

```python
x = []
ue_x = []
def u(x):
    return 3*np.exp(-x) + x - 2
```

```python
for i in range(N):
    x.append(a + (i*dx))
    ue_x.append(u(x[i]))
```

```python
plt.plot(x, ue_x, '-')
plt.grid()
```

```python
plt.plot(x, ue_x, '-')
plt.grid()

u_num = []
u_num.append(1)
for i in range(1, N):
    approx = (1 / (1 + dx)) * (u_num[i-1] + dx*(x[i] - 1))
    u_num.append(approx)
print(u_num)
plt.plot(x, u_num, "o-r")
plt.grid()
```

```python
# axecte
plt.plot(x, ue_x, '-')
plt.grid()

# approx decentr
u_num = []
u_num.append(1)
for i in range(1, N):
    approx = (1 / (1 + dx)) * (u_num[i-1] + dx*(x[i] - 1))
    u_num.append(approx)
print(u_num)
plt.plot(x, u_num, "-r")
plt.grid()

#approx excentr
u_num2 = []
u_num2.append(1)
for i in range(N-1):
    approx = u_num2[i]*(1 - dx) + dx*(x[i] - 1)
    u_num2.append(approx)
print(u_num2)
plt.plot(x, u_num2, "-g")
plt.grid()

plt.axis([0.6, 1.25, 0, 0.5])
```

```python
# Function 
def u(x):
    return 3*np.exp(-x) + x - 2

# Interval
a = 0
b = 2
N = 100
```

```python
# Approximation
def diff_finies(a, b, N, u):
    # axe
    x = np.linspace(a, b, N)
    dx = (b-a) * (N-1)

    # u experimental
    ue_x = [u(e) for e in x]
    plt.plot(x, ue_x, 'o-')
    plt.grid()
    print(ue_x)

    # u numeric
    u_num = []
    u_num.append(1)
    for i in range(1, N):
        approx = (1 / (1 + dx)) * (u_num[i-1] + dx*(x[i] - 1))
        u_num.append(approx)
    print(u_num)
    plt.plot(x, u_num, "o-r")
    plt.grid()
```

```python
diff_finies(a, b, N, u)
```
