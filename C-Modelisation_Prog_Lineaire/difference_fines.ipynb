```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0
b = 2
N = 90
```

```python
dx = (b - a) / (N-1)
```

```python
x = []
ue_x = []
def u(x):
    return 3*np.exp(-x) + x - 2
```

```python
for i in range(N):
    x.append(a + (i*dx))
    ue_x.append(u(x[i]))
```

```python
plt.plot(x, ue_x, 'o-')
plt.grid()
```

```python
u_num = []
u_num.append(1)
for i in range(1, N):
    approx = (1 / (1 + dx)) * (u_num[i-1] + dx*(x[i] - 1))
    u_num.append(approx)
plt.plot(x, u_num, "o-r")
plt.grid()
```

Version avancé

```python
# Function 
def u(x):
    return 3*np.exp(-x) + x - 2

# Interval
a = 0
b = 2
N = 90
```

```python
# Approximation
def diff_finies(a, b, N, u):
    x = np.linspace(a, b, N)
    dx = (b-a) * (N-1)
    u_numeric = [0]
    for i in range(1, N):
        approx = (1 / (1 + dx)) * (u_numeric[i-1] + dx*(x[i] - 1))
        u_numeric.append(approx)
    
    plt.plot(x, u_numeric, "o-r")
    plt.grid()
```

```python
diff_finies(a, b, N, u)
```
