```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Condition initiale
def f(x):
    if (300 <= x <= 400):
        return 10
    return 0
```

```python
# Dimenssions(Longeur) de la barre
L = 1000

# Nombre de neuds
N = 100

# Pas du maillage
dx = L / (N-1)

# Maillage de la barre
x = np.linspace(0, L, N)

# Conditions limites
U = np.array([f(x[i]) for i in range(N)])
```

```python
plt.plot(x, U, '-b')
plt.grid()
```

```python
a = 1
temps = 0
tempsArret = 600
CFL = 0.8

dt = CFL * (dx/abs(a))
lamba = dt * (a/dx)

Unew = np.zeros(N)
```

```python
k = 0
while (temps < tempsArret):
    for i in range(1, N-1):
        Unew[i] = U[i] - lamba*(U[i] - U[i-1])
    Unew[0] = 0
    Unew[N-1] = 0
    temps += dt
    U = Unew.copy()
    plt.figure()
    plt.plot(x, U, '-r')
plt.grid()
```
