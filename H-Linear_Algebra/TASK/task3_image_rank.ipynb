```python
#import
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load image
image_path="./image/forest.png"
image_path="./image/flower.png"
image = Image.open(image_path)
image_gray = image.convert("L")

# Convert the image to the nupy array
image_array = np.array(image_gray)

# Perform singular Value Decomposition
U, S, V = np.linalg.svd(image_gray)

# Set the rank for the low-rank appriximation
rank = 50

# Reconstrcut the image with rank approximation
approximation = np.dot(U[:, :rank], np.dot(np.diag(S[:rank]), V[:rank, :]))
```

```python
# Original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_array, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(approximation, cmap='gray')
plt.title('Approximation')
plt.axis('off')

plt.show()
```

```python
# Compare difference
diff  = image_array-approximation
error = np.linalg.norm(diff, 2)
print(f"error\t : {error}")
sigma_k1 = S[rank]
print(f"sigma_k1 : {sigma_k1}")
```
