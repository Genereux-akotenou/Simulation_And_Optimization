```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a gamma-squared random variable
def generate_gamma_squared(n_samples, shape):
    gamma_samples = np.random.gamma(shape, scale=1, size=(n_samples,))
    gamma_squared_samples = gamma_samples**2
    return gamma_squared_samples

# Function to visualize the Central Limit Theorem
def visualize_clt(samples, num_subplots=4):
    plt.figure(figsize=(12, 8))
    
    # Original distribution
    plt.subplot(num_subplots, 1, 1)
    plt.hist(samples, bins=50, density=True, color='blue', alpha=0.7)
    plt.title('Histogram of Gamma-squared Random Variable')
    
    # CLT for different sample sizes
    for i in range(2, num_subplots + 1):
        plt.subplot(num_subplots, 1, i)
        sample_size = 2**i
        means = [np.mean(np.random.choice(samples, size=sample_size)) for _ in range(1000)]
        plt.hist(means, bins=50, density=True, color='green', alpha=0.7)
        plt.title(f'CLT with Sample Size {sample_size}')

    plt.tight_layout()
    plt.show()

# Generate gamma-squared random variable
n_samples = 10000
shape_parameter = 2
gamma_squared_samples = generate_gamma_squared(n_samples, shape_parameter)

# Visualize the Central Limit Theorem
visualize_clt(gamma_squared_samples)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_gamma_squared(a, b, n_samples):
    return np.random.gamma(a, scale=b, size=n_samples)**2

def clt_visualization(a, b, sample_sizes):
    plt.figure(figsize=(12, 8))

    for i, sample_size in enumerate(sample_sizes, 1):
        # Generate gamma squared random variables
        gamma_squared_samples = generate_gamma_squared(a, b, sample_size)

        # Calculate sample means
        sample_means = [np.mean(np.random.choice(gamma_squared_samples, size=sample_size)) for _ in range(1000)]

        # Plot histograms
        plt.subplot(2, 2, i)
        plt.hist(sample_means, bins=50, density=True, alpha=0.7)
        plt.title(f'Sample Size = {sample_size}')

    plt.tight_layout()
    plt.show()

# Visualization
a_param = 2
b_param = 4
sample_sizes_to_test = [10, 50, 100, 500]
clt_visualization(a_param, b_param, sample_sizes_to_test)
```
