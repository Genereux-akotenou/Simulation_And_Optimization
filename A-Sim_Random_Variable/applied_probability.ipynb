<b>Variables aléatoires</b>: 
Une variable aléatoire est une fonction qui associe des valeurs numériques aux résultats d'une expérience aléatoire.

<ul>
    <li>
        Variable Aléatoire Discrète
        <ul>
            <li>- Prend des valeurs distinctes et isolées.</li>
            <li>- Exemples : le nombre de faces d'un dé après un lancer, le nombre d'étudiants dans une classe.</li>
            <li>- Fonction de probabilité : assigne une probabilité à chaque valeur possible, la somme de toutes les probabilités étant égale à 1.</li>
        </ul>
    </li>
    <li>
        Variable Aléatoire Continue
        <ul>
            <li>- Peut prendre n'importe quelle valeur dans un intervalle.</li>
            <li>- Exemples : la taille, le poids.</li>
            <li> - Caractérisée par une **fonction de densité de probabilité** (PDF) au lieu d'une fonction de probabilité.</li>
       <ul>
    </li>
</ul>
</div>

##### 1 - Variable Aléatoire Discrète $ X $

- Fonction de Probabilité
    - Pour tout $ x $, $ P(X = x) >= 0 $
    - $ \sum_{x}^{} P(X = x) = 1 $

- Fonction de Probabilité Cumulative (CDF) :
    - $ F(x) = P(X < x) $ 
    - F est non décroissante 

- Expérance $ E[X] $
    - La moyenne pondérée des valeurs possibles de X
    - $ E[X] = \sum_{x}^{} x.P(X = x) $

- Variance $ Var[X]) $
    - Mesure de la dispersion des valeurs de X autour de son espérance
    - $ Var[X]=E[(X − E[X])^2] $
    
- Écart-type $ σ(X) $
    - Mesure de dispersion qui a la même unité que X.
    - $ σ(X) = \sqrt{Var[X]} $

##### 2 - Variable Aléatoire Continue $ X $

- Fonction de Densité de Probabilité (PDF)
    - Pour tout $ x $, $ f(x) \geq 0 $
    - $ \int_{-\infty}^{\infty} f(x) \,dx = 1 $

- Fonction de Répartition Cumulative (CDF)
    - $ F(x) = P(X \leq x) $ 
    - F est non-décroissante 
    - $ \lim_{{x \to -\infty}} F(x) = 0 $ et $ \lim_{{x \to \infty}} F(x) = 1 $

- Espérance $ E[X] $
    - La moyenne pondérée des valeurs possibles de X
    - $ E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \,dx $

- Variance $ \text{Var}[X] $
    - Mesure de la dispersion des valeurs de X autour de son espérance
    - $ \text{Var}[X] = \int_{-\infty}^{\infty} (x - E[X])^2 \cdot f(x) \,dx $

- Écart-type $ \sigma(X) $
    - Mesure de dispersion qui a la même unité que X.
    - $ \sigma(X) = \sqrt{\text{Var}[X]} $

3 - Variable aleatoire uniforme
<hr>

##### - Généralité

Une variable aléatoire X est dite suit la loi uniforme sur ]0,1[ si sa fonction de densité de probabilité fX est définie par :

$$ 
f_X(t) =
\begin{cases} 
1 & \text{si } 0 < t < 1 \\
0 & \text{sinon}
\end{cases}
$$

Ou bien, sa fonction de répartition \(F_X\) est définie par :

$$
F_X(t) =
\begin{cases} 
1 & \text{si } t \geq 1 \\
t & \text{si } 0 < t < 1 \\
0 & \text{si } t \leq 0
\end{cases}
$$

Plus généralement, la variable aléatoire \(X\) est dite suivre la loi uniforme sur \([a, b[\) si sa fonction de densité de probabilité \(f_X\) est définie par :

$$
f_X(t) =
\begin{cases} 
\frac{1}{b-a} & \text{si } a < t < b \\
0 & \text{sinon}
\end{cases}
$$

Et la fonction de répartition \(F_X\) devient :

$$
F_X(t) =
\begin{cases} 
0 & \text{si } t \leq a \\
\frac{t - a}{b - a} & \text{si } a < t < b \\
1 & \text{si } t \geq b
\end{cases}
$$

##### - Propriétés
Si $ U\leadsto Uniforme(0, 1) $ et $ Z\leadsto Uniforme(a, b) $ alors une ecriture de Z est: $ Z = (b - a)U + a $

Réaliser la simulation de la loi de Bernoulli de paramètre p et en déduire la simulation de la loi Binomiale de paramètres(n, p) et la simulation de la loi géométrique de paramètre p

- loi de Bernoulli: Une variable aléatoire suivant la loi de Bernoulli est appelée variable de Bernoulli. Plus formellement, une variable aléatoire X suit la loi de Bernoulli de probabilité p si
<br>
<br>
$ {\displaystyle \mathbb {P} (X=x)=p^{x}(1-p)^{1-x},x\in \{0;1\}} $ 

```python
def simulation_bernouilli(p):
    u = np.random.uniform(0, 1)
    if u < p:
        return 1
    return 0
```

- Loi binomiale: la loi binomiale modélise la fréquence du nombre de succès obtenus lors de la répétition de plusieurs expériences aléatoires identiques et indépendantes
<br>
On a:
<br><br>
$ n\in \mathbb {N} $, 
$ {\displaystyle p\in [0,1]} $, 
$ q=1-p $
<br>
$ P(X=k) = {n \choose k}p^{k}q^{{n-k}}\! $

```python
def simulation_binomiale(n, p):
    sum = 0
    for i in range(n):
        sum += simulation_bernouilli(p)
    return sum
```

- Loi géométrique: la loi géométrique désigne, selon la convention choisie, l'une des deux lois de probabilité suivantes :
     - la loi du nombre X d'épreuves de Bernoulli indépendantes de probabilité de succès p ∈ ]0,1[ nécessaire pour obtenir le premier succès.
     - La loi du nombre Y = X – 1 d'échecs avant le premier succès.
<br>
<br>
$ {\displaystyle \mathbb {P} (X=k)=q^{k-1}p.} $

```python
def simulation_geometrique(p):
    try_without_success = 0
    while simulation_bernouilli(p) == 0:
        try_without_success += 1
    return try_without_success+1
```

<br>
<br>
**MATH**: 
<br>
1 - La fonction de densité de probabilité de X est $ {\displaystyle \lambda \mathrm {e} ^{-\lambda x}} $ et sa fonction de repartition est donné par: 
$ {\displaystyle 1-\mathrm {e} ^{-\lambda x}}Ind(x>0) $. 
<br>
2 - FX est un bijection de ]0, ∞[−→]0, 1[. Posons Fx = y
<br>
Fx = y  <=> 1 - e-λx = y<br>

```python
def simulation_exponentiel(lambda_):
    u = np.random.uniform(0, 1)
    return (-np.log(1 - u)) / lambda_
```

- Montrer que Y suit la loi géométrique et déterminer son paramètre en fonction de λ
<br>
<br>
**MATH**: 
<br>
Y est une varaible aleatoire discrete telle que Y = int(X) + 1.
<br>
Calculons: <br>
\begin{align*}
P(Y=k) & = P(\lfloor X \rfloor + 1 = k) \\
       & = P(\lfloor X \rfloor = k - 1) \\
       & = P(k - 1 \leq x < k) \\
       & = \int_{k-1}^{k}f_X(x)dx \\
       & = \int_{k-1}^{k}\lambda\exp^{-\lambda x}dx \\
       & = \exp^{-\lambda (k-1)}(1 - \exp^{\lambda}) \\
\end{align*}

Ainsi : $ Y \leadsto Geometrique(p),   p=1 - \exp^{\lambda} $

- En déduire une méthode pour simuler la loi géométrique de paramètre p

```python
def simulation_geometrique_y(p):
    lambda_ = -np.log(1 - p) 
    return np.floor(simulation_exponentiel(lambda_)) + 1
```

```python
def surfaceB():
    n = 100000
    sum = 0
    for i in range(n):
        sum += 2*np.abs(np.sin(sim_x()))
    return (1/n) * sum
```

```python
def Normal_01():
    """
    We use box_muller to simulate Normal(0, 1)
    """
    u = np.random.uniform(0, 1)
    v = np.random.uniform(0, 1)
    return np.sqrt(-2 * np.log(u)) * np.sin(2 * np.pi * v)

def Normal(m, p):
    return p * Normal_01() + m
```

```python
h = 0.01
t = np.arange(0, 100, h)
f_xt_array = [0]

for i in t[1:]:
    f_xt_array.append(f_xt_array[-1] + Normal(0, np.sqrt(h)))

plt.plot(t, f_xt_array, label='(t, Xt)', color="blue")
plt.legend()
```

- Tracer la courbe paramétrique (Xt,Yt)

```python
h = 0.01
t = np.arange(0, 100, h)
f_xt_array = [0]
f_yt_array = [0]

for i in t[1:]:
    f_xt_array.append(f_xt_array[-1] + Normal(0, np.sqrt(h)))

for i in t[1:]:
    f_yt_array.append(f_yt_array[-1] + Normal(0, np.sqrt(h)))

plt.plot(f_xt_array, f_yt_array, label='(Xt, Yt)', color="red")
plt.legend()
```

```python
def N(t, l):
    n = 0
    Sn = 0
    while True:
        Sn += Xn(l)
        n += 1
        if Sn > t:
            break
    return n
```

```python
def Sn(n, lambda_):
    Sn_times = []
    S = 0
    for i in range(n):
        x = Xn(lambda_)
        S += x
        Sn_times.append(S)
    return Sn_times
```
