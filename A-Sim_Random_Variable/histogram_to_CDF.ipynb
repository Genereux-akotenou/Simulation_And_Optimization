```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
print(data)

# Create a histogram
plt.hist(data, bins=30, density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the histogram
plt.show()

# Create cumulative density function (CDF)
sorted_data = np.sort(data)
cumulative = np.arange(len(sorted_data)) / float(len(sorted_data))
print(cumulative)

# Plot the CDF
plt.plot(sorted_data, cumulative, color='red')
plt.title('Cumulative Density Function (CDF)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')

# Display the CDF graph
plt.show()
```
