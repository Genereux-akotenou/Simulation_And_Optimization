```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm, uniform

# np.random.seed(0)

# Generate dataa
df = 4
chi2_data = np.random.chisquare(df, size=200)

# Create QQ plot
sm.qqplot(chi2_data, dist=uniform, line='45', fit=True)
plt.title('QQ Plot - Chi-Squared Distribution')
plt.show()
```
